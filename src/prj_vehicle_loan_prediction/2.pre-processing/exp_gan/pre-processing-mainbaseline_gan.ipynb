# 4.Chuẩn hóa dữ liệu

```python
numeric_columns=dict(np.load(f'{exps_dir}/exp_prj_vehicle_loan_prediction/columns_dtype.npz',allow_pickle=True))['numeric_columns']
numeric_columns=numeric_columns[1:]
numeric_columns
```

```python
category_columns=dict(np.load(f'{exps_dir}/exp_prj_vehicle_loan_prediction/columns_dtype.npz',allow_pickle=True))['category_columns']
category_columns
```

```python
data_train.LOAN_DEFAULT
```

```python
label_encoders = {}
for column in category_columns:
    label_encoder = LabelEncoder()
    data_train[column] = label_encoder.fit_transform(data_train[column])
    label_encoders[column] = label_encoder  # Lưu trữ label encoder nếu cần sau này

scaler = StandardScaler()
data_train[numeric_columns] = scaler.fit_transform(data_train[numeric_columns])
data_train
```

```python
# from imblearn.over_sampling import SMOTE
# x_train, y_train = SMOTE().fit_resample(x_train, y_train)
from tabgan.sampler import GANGenerator

gen_x, gen_y = GANGenerator(gen_x_times=10.5, 
                                cat_cols=None,
                                bot_filter_quantile=0.01, 
                                top_filter_quantile=0.99,
                                is_post_process=True,
                                adversarial_model_params={
                                    "metrics": "accuracy", 
                                    "max_depth": 2, 
                                    "max_bin": 100, 
                                    "learning_rate": 0.02, 
                                    "random_state": 42, 
                                    "n_estimators": 500,
                                },
                                pregeneration_frac=2,
                                only_generated_data=False,
                                gen_params = {
                                    "batch_size": 500, 
                                    "patience": 25, 
                                    "epochs" : 500,
                                }).generate_data_pipe(data_train.drop('LOAN_DEFAULT',axis=1), data_train['LOAN_DEFAULT'].to_frame(), data_test.drop('LOAN_DEFAULT',axis=1), 
                                                      deep_copy=True, only_adversarial=False, 
                                                      use_adversarial=True)

gen_x.to_excel(f'{save_dir}/gen_x.xlsx', index=False)
gen_y.to_excel(f'{save_dir}/gen_y.xlsx', index=False)
```

```python
df_gen=gen_x
df_gen['LOAN_DEFAULT']=gen_y
df_train=pd.concat([data_train,df_gen[df_gen.LOAN_DEFAULT==1]] , axis=0, ignore_index=True)
```

```python
x_train=df_train.drop('LOAN_DEFAULT',axis=1)
y_train=df_train['LOAN_DEFAULT']
x_train.to_csv(f'{save_dir}/x_train.csv', index=False)
y_train.to_csv(f'{save_dir}/y_train.csv', index=False)
```
